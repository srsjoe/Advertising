```python
#Python Code for Needleman-Wunsch (NW) Scoring Matrix: 

#Code: 

#Any two DNA sequences and scoring rules, e.g., m = 1, mm = 0, g = -1.  

m=1  #match 

mm=0 #mismatch 

g=-1 #gap  

# Function for a matrix of zeroes 

def zeros(rows, cols): 

    e_list = [] #  empty list 

    # Matrix Setup 

    for x in range(rows): 

        e_list.append([]) # Empty list added in each row 

        for y in range(cols): # Columns added in each row 

            e_list[-1].append(0) # Adds zero to each column and row 

    return e_list # Retruns the matrix of zeros 

def max_score(x,y): # A function for that compares match, mismatch and gap in an alignment 

    if x == y: 

        return m 

    elif x == '-' or y == '-': 

        return g 

    else: 

        return mm 

def NW_Matrix(seq_1, seq_2): # This function adds scores to the matrix when two sequenses are given(seq1 and seq2) 

     

    l1 = len(seq_1) # length of seq1 

    l2 = len(seq_2) # length of seq2 

     

    score = zeros(l1+1, l2+1) # A placeholder for storing scores that generates a matrix of zeros. 

     

    for i in range(0, l1 + 1): # scores first column 

        score[i][0] = g * i 

     

    for j in range(0, l2 + 1): # scores first row 

        score[0][j] = g * j 

     

    for i in range(1, l1 + 1): # Socring all other values in the matrix  

        for j in range(1, l2 + 1): 

            # Checking scores form diagonal,horizntal and vertical cells 

            d = score[i - 1][j - 1] + max_score(seq_1[j-1], seq_2[i-1]) 

            h = score[i - 1][j] + g 

            v = score[i][j - 1] + g 

            score[i][j] = max(d, h, v) # Maximum score from calculating diagolal,horizontal and vertical values 

    return score 

# NW Matrix with correct scores between two sequences 

seq_1 = "ATC" 

seq_2 = "ATC" 

M = NW_Matrix(seq_1, seq_2)  

  

for row in M:  

  

    for column in row:  

  

        print(column, end=" ")  

  

    print('', sep='\n') #Generates NW Matrix for two given sequences.
```
